```python
#Import all necessery libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#ML
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
```

```python
pd.set_option("display.max_rows",None)
pd.set_option("display.max_columns",None)
```

```python
# read dataset
train = pd.read_csv('input/train.csv')
test = pd.read_csv('input/test.csv')
test_ans = pd.read_csv('input/sample_submission.csv')
```

```python
columns = []
for column,column_type in zip(train.columns,train.dtypes):
    if column_type in ['int64','float64']:
        columns.append(column)
columns.pop(0)
```

```python
train.describe()
```

```python
#Linear Regression Algorithm
lr = LinearRegression()
# Random Forest Algorithm (1000 decision trees)
rf = RandomForestRegressor(n_estimators = 1000, random_state = 0)
```

```python
train = train[columns]
test = test[columns[:-1]]
test['SalePrice'] = test_ans['SalePrice']
```

```python
train.dropna(inplace=True)
test.dropna(inplace=True)
```

```python
x_train = train.drop(['SalePrice'], axis=1)
y_train = train['SalePrice']

x_test = test.drop(['SalePrice'], axis=1)
y_test = test['SalePrice']
```

```python
model_lr = lr.fit(x_train, y_train)
y_pred = lr.predict(x_test)
```

```python
print(y_pred)
```

```python
Rpow2 = model_lr.score(x_train,y_train)

b0 = model_lr.intercept_
b1 = model_lr.coef_

y_pred = model_lr.predict(x_test)

df = pd.DataFrame({
    'Actual':y_test,
    'Predict':y_pred,
    'Errors':y_test-y_pred
})

print("R^2=",Rpow2,"\nbo=",b0,"\nb1, b2, ... =",b1)
print(df)
```

```python
model_rf = rf.fit(x_train, y_train)
y_pred = rf.predict(x_test)
```

```python
Rpow2 = model_rf.score(x_train,y_train)

y_pred = model_rf.predict(x_test)

df = pd.DataFrame({
    'Actual':y_test,
    'Predict':y_pred,
    'Errors':y_test-y_pred
})

print("R^2=",Rpow2)
print(df)
```
